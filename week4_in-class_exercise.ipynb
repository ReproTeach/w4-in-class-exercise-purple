## 2. Preprocess the data (from week3 exercise)

1. Rename all columns
2. Drop nan
3. Drop duplicates

```python
def process_data(filename):
    
    csv_file="pre-course_survey.csv"
    df_csv=pd.read_csv(csv_file)
    
    old_cols = df_csv.columns
    new_cols = ["time", "python", "concepts", "git", "jupyter", "repro", "nlp", "social_media", "goals", "skills", "challenges"]
    cols_dict = dict(zip(old_cols, new_cols))
    df_csv.rename(columns=cols_dict, inplace=True)
    
    df_csv.dropna(inplace=True)
    df_csv[df_csv.duplicated()]
    df_csv.drop_duplicates()
    
    df_csv1 = df_csv.replace({"Yes":1,"No":0})

    return df_csv1
```

```python
process_data(csv_file)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data=process_data(csv_file)
```

### Line Plot

```python
sns.set_style("whitegrid")
sns.set_palette("pastel")
sns.lineplot(x="python", y="repro", color='purple',marker='o', data=data)
plt.xticks([0, .5, 1, 1.5, 2, 2.5, 3, 3.5, 4,])
plt.yticks([0, .5, 1, 1.5, 2, 2.5, 3, 3.5, 4,])
plt.title('Python and Reproducible Research')
plt.xlabel('Knowledge of Python')
plt.ylabel('Familiarity with Reproducible Research')
plt.show()
```

### Scatter Plot 

```python
data=process_data(csv_file)
sns.set_style("whitegrid")
sns.scatterplot(x="jupyter", y="repro", color= 'purple', data=data)
plt.xticks([0, .5, 1,])
plt.title('Jupyter and Reproducible Research')
plt.xlabel('Familiarity with Jupyter')
plt.ylabel('Familiarity with Reproducible Research')
plt.show()
```

### Violin Plot 

```python
sns.violinplot(x = "git", y ="python", color= 'purple', data = data,)
plt.title('Python and Git')
plt.xlabel('Familiarity with Git')
plt.ylabel('Familiarity with Python')
plt.show()
```
